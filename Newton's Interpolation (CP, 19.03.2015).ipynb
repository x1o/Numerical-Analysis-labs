<h1>Newton's Interpolation</h1>

```python
from pprint import pprint

PREC = 7

# Graph
G = (
(0,   1.614419),
(0.1, 1.656832),
(0.2, 1.694888),
(0.3, 1.728606),
(0.4, 1.758030),
(0.5, 1.783225),
(0.6, 1.804279),
(0.7, 1.821299),
(0.8, 1.834414),
(0.9, 1.843768)
)

# Table of finite differences
TFD = []
ORDER = 4
# FIXME: not very efficient
X = [x.n(digits=PREC) for (x, y) in G]
Y = [y.n(digits=PREC) for (x, y) in G]
TFD.append(Y)
for _ in range(ORDER):
    TFD.append([(TFD[-1][i+1] - TFD[-1][i]) for i in range(len(TFD[-1])-1)])
#pprint(TFD)
#pprint([X] + TFD)
html(table([X] + TFD,
                header_row=range(len(G)),
                header_column=['', '$x', '$y']+['$\Delta^%s$' % i for i in range(1,ORDER+1)]))
```

```python
var('t')
h = 0.1
x1s = 0.105582
closest_idx = floor(x1s / h)
html('Index of the value of $x$ closest to $x^*_1$: $$%s$$' % closest_idx)
DY = [col[closest_idx] for col in TFD]
html('Choosing $\Delta^j y_%s, j\in 0:%s$:' % (closest_idx, ORDER))
show(DY)
P(t) = sum(DY[i] * prod(t-j for j in range(i)) / factorial(i) for i in range(ORDER+1))
html('Newton polynomial $P(t)$:')
show(P)
html('Simplified:')
show(P.simplify_full())
html(r'$P(\frac{x_1^* - x_0}{h})$:')
show(P((x1s - X[closest_idx])/h))
```
